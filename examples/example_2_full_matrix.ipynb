```python
import numpy as np
from matrix_lib import FullMatrix
```

```python
zeros_int = FullMatrix.zero(4, 4, default=0) # matrix with all 0 elements of type int64
zeros_flt = FullMatrix.zero(2, 4, default=0.) # matrix with all 0 elements of type float64
tens_flt = FullMatrix.zero(3, 2, default=10.) # matrix with all 10 elements of type float64
```

```python
zeros_int[0, 2] = -1 # assign element
zeros_flt[1, 1] = 3.14
tens_flt[2, 1] = -10.0
print(zeros_int[0, 2]) # get element
print(zeros_flt[1, 1])
```

```python
print("zeroes_int =") # printing matrices
print(zeros_int)
print("zeroes_flt =")
print(zeros_flt)
print("tens_flt =")
print(tens_flt)
```

```python
# sum and diff
ones_int = FullMatrix.zero(zeros_int.shape[0], zeros_int.shape[1], default=1)
sum_matrix = ones_int + zeros_int
print(sum_matrix)
```

```python
# matmul and inverse
mt1_flt = FullMatrix(data=np.array([[1, -1, 3], [2, 5, -1], [1, 1, -1]], dtype=np.float64))
mt2_flt = FullMatrix(data=np.array([[-2, -5, 0], [0.5, -3, 1], [-0.5, 4, 4]], dtype=np.float64))
print(mt1_flt@mt2_flt)
print(mt1_flt.inverse())
print(mt2_flt.inverse())
```
